# Day 6: Universal Orbit Map

```python
EXAMPLE = """COM)B
B)C
C)D
D)E
E)F
B)G
G)H
D)I
E)J
J)K
K)L""".split()

def parse(lines):
    orbits = dict()
    for line in lines:
        c,o = line.split(")")
        orbits[o] = c
    return orbits
print(parse(EXAMPLE))
```

```python
def trace(o, orbits):
    while True:
        if not o in orbits:
            return
        o = orbits[o]
        yield o
parsed = parse(EXAMPLE)
print(list(trace('D', parsed)))
print(list(trace('L', parsed)))
print(list(trace('COM', parsed)))
```

```python
def checksum(orbits):
    cs = 0
    for o in orbits.keys():
        cs += len(list(trace(o, orbits)))
    return cs
print(checksum(parsed))
```
